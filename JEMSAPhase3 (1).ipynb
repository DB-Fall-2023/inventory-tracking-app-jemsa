```python
import ipywidgets as widgets
from IPython.display import display
import requests
import matplotlib.pyplot as plt
import json
from IPython.display import clear_output
import numpy
```

```python
def on_most_incoming_global_stats_button_click(button):

  #Get Most Incoming
  most_incoming_url = "https://postgres-sql-6aba726b2968.herokuapp.com/jemsa/most/incoming"
  most_incoming_response = requests.get(most_incoming_url)
  most_incoming = most_incoming_response.json()

  # Clear old plot
  clear_output(wait=True)

  # Redisplay the buttons
  display(fetch_most_incoming_global_stats_button)

  #Plot the graph
  warehouse_names = [entry['WarehouseName'] for entry in most_incoming]
  transaction_counts = [entry['Count of Incoming Transactions per Warehouse'] for entry in most_incoming]

  plt.bar(warehouse_names, transaction_counts)
  plt.xlabel('Warehouse Name')
  plt.ylabel('Count of Incoming Transactions')
  plt.title('Top 5 Warehouses based on Incoming Transactions')

#Button to fetch top 5 most incoming transactions warehouses
fetch_most_incoming_global_stats_button = widgets.Button(
    description = 'Fetch Top 5 Warehouses by Incoming Transactions',
    layout = widgets.Layout(width = '400px')
)
fetch_most_incoming_global_stats_button.on_click(on_most_incoming_global_stats_button_click)

# Display buttons

display(fetch_most_incoming_global_stats_button)
```

```python
def on_most_racks_global_stats_button_click(button):

  #Get Top 10 Warehouses with the Most Racks
  most_racks_url = "https://postgres-sql-6aba726b2968.herokuapp.com/jemsa/most/rack"
  most_racks_response = requests.get(most_racks_url)
  most_racks = most_racks_response.json()

  # Clear old plot
  clear_output(wait=True)

  # Redisplay the buttons
  display(fetch_most_racks_global_stats_button)

  #Plot the graph
  warehouse_names = [entry['WarehouseName'] for entry in most_racks]
  transaction_counts = [entry['Count of Racks per Warehouse'] for entry in most_racks]

  # Create a pie chart
  plt.pie(transaction_counts, labels=warehouse_names, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'purple', 'orange'])
  # Add a title
  plt.title('Top 10 Warehouses based on Racks per Warehouse Transactions')
  # Display the pie chart
  plt.show()

#Button to fetch top 5 most racks warehouses
fetch_most_racks_global_stats_button = widgets.Button(
    description = 'Fetch Top 10 Warehouses by Racks',
    layout = widgets.Layout(width = '400px')
)
fetch_most_racks_global_stats_button.on_click(on_most_racks_global_stats_button_click)

# Display buttons

display(fetch_most_racks_global_stats_button)
```
